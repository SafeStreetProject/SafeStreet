```python
import torch
import torch.nn as nn
from torchvision import datasets, transforms
from torch.utils.data import DataLoader
from PIL import Image

# Step 1: Define your custom ViT
class SimpleViT(nn.Module):
    def __init__(self, image_size=224, patch_size=16, num_classes=2, dim=768, depth=6, heads=8, mlp_dim=2048):
        super(SimpleViT, self).__init__()
        num_patches = (image_size // patch_size) ** 2
        self.patch_embed = nn.Conv2d(3, dim, kernel_size=patch_size, stride=patch_size)
        self.transformer = nn.TransformerEncoder(
            nn.TransformerEncoderLayer(d_model=dim, nhead=heads, dim_feedforward=mlp_dim),
            num_layers=depth
        )
        self.classifier = nn.Linear(dim, num_classes)

    def forward(self, x):
        x = self.patch_embed(x)
        x = x.flatten(2).transpose(1, 2)
        x = self.transformer(x)
        x = x.mean(dim=1)  # Global average pooling
        return self.classifier(x)

# Step 2: Mount Google Drive
from google.colab import drive
drive.mount('/content/drive')  # Already mounted, no force_remount unless needed

# Step 3: Define transformations
transform = transforms.Compose([
    transforms.Resize((224, 224)),
    transforms.ToTensor(),
    transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225])
])

# Step 4: Load dataset with error checking
try:
    train_dataset = datasets.ImageFolder(root="/content/drive/MyDrive/programs/dataset/train", transform=transform)
    print("Train dataset loaded successfully:", len(train_dataset), "images")
    val_dataset = datasets.ImageFolder(root="/content/drive/MyDrive/programs/dataset/val", transform=transform)
    print("Validation dataset loaded successfully:", len(val_dataset), "images")
except FileNotFoundError as e:
    print(f"Dataset loading failed: {e}")
    raise

train_loader = DataLoader(train_dataset, batch_size=4, shuffle=True)
val_loader = DataLoader(val_dataset, batch_size=4)

# Step 5: Initialize and train the ViT
model = SimpleViT().to("cuda")  # Use Tesla T4 GPU
optimizer = torch.optim.Adam(model.parameters(), lr=0.001)
criterion = nn.CrossEntropyLoss()

print("Starting initial training for 1 epoch...")
for epoch in range(1):  # Initial training
    model.train()
    for images, labels in train_loader:
        images, labels = images.to("cuda"), labels.to("cuda")
        optimizer.zero_grad()
        outputs = model(images)
        loss = criterion(outputs, labels)
        loss.backward()
        optimizer.step()
print(f"Initial training loss: {loss.item()}")

# Save the initial model
torch.save(model.state_dict(), "/content/drive/MyDrive/programs/road_damage_detection_backup/best.pt")
print("Initial model saved as best.pt")

# Step 6: Fine-tune the model
optimizer = torch.optim.Adam(model.parameters(), lr=0.0001)  # Lower LR for fine-tuning
print("Starting fine-tuning for 3 epochs...")
for epoch in range(3):
    model.train()
    for images, labels in train_loader:
        images, labels = images.to("cuda"), labels.to("cuda")
        optimizer.zero_grad()
        outputs = model(images)
        loss = criterion(outputs, labels)
        loss.backward()
        optimizer.step()
    print(f"Fine-tuning Epoch {epoch+1}/3, Loss: {loss.item()}")

    # Validation
    model.eval()
    correct = 0
    total = 0
    with torch.no_grad():
        for images, labels in val_loader:
            images, labels = images.to("cuda"), labels.to("cuda")
            outputs = model(images)
            _, predicted = torch.max(outputs, 1)
            total += labels.size(0)
            correct += (predicted == labels).sum().item()
    print(f"Validation Accuracy: {100 * correct / total}%")

# Save the fine-tuned model
torch.save(model.state_dict(), "/content/drive/MyDrive/programs/road_damage_detection_backup/fine_tuned_best.pt")
print("Fine-tuned model saved as fine_tuned_best.pt")
```

```python
import torch
import torch.nn as nn
from torchvision import datasets, transforms
from torch.utils.data import DataLoader
from PIL import Image

# Step 1: Define your custom ViT with batch_first=True
class SimpleViT(nn.Module):
    def __init__(self, image_size=224, patch_size=16, num_classes=2, dim=768, depth=6, heads=8, mlp_dim=2048):
        super(SimpleViT, self).__init__()
        num_patches = (image_size // patch_size) ** 2
        self.patch_embed = nn.Conv2d(3, dim, kernel_size=patch_size, stride=patch_size)
        self.transformer = nn.TransformerEncoder(
            nn.TransformerEncoderLayer(d_model=dim, nhead=heads, dim_feedforward=mlp_dim, batch_first=True),
            num_layers=depth
        )
        self.classifier = nn.Linear(dim, num_classes)

    def forward(self, x):
        x = self.patch_embed(x)  # (batch, dim, patches_h, patches_w)
        x = x.flatten(2).transpose(1, 2)  # (batch, num_patches, dim)
        x = self.transformer(x)  # (batch, num_patches, dim) with batch_first=True
        x = x.mean(dim=1)  # Global average pooling
        return self.classifier(x)

# Step 2: Mount Google Drive
from google.colab import drive
drive.mount('/content/drive')

# Step 3: Define transformations with augmentation
transform = transforms.Compose([
    transforms.Resize((224, 224)),
    transforms.RandomHorizontalFlip(),
    transforms.RandomRotation(10),
    transforms.ToTensor(),
    transforms.Normalize(mean=[0.485, 0.456, 0.406], std=[0.229, 0.224, 0.225])
])

# Step 4: Load dataset
try:
    train_dataset = datasets.ImageFolder(root="/content/drive/MyDrive/programs/dataset/train", transform=transform)
    print("Train dataset loaded successfully:", len(train_dataset), "images")
    val_dataset = datasets.ImageFolder(root="/content/drive/MyDrive/programs/dataset/val", transform=transform)
    print("Validation dataset loaded successfully:", len(val_dataset), "images")
except FileNotFoundError as e:
    print(f"Dataset loading failed: {e}")
    raise

train_loader = DataLoader(train_dataset, batch_size=8, shuffle=True)  # Adjust batch size based on data size
val_loader = DataLoader(val_dataset, batch_size=8)

# Step 5: Load existing model and fine-tune
model = SimpleViT().to("cuda")  # Use Tesla T4 GPU
model.load_state_dict(torch.load("/content/drive/MyDrive/programs/road_damage_detection_backup/best.pt", map_location="cuda"))
optimizer = torch.optim.Adam(model.parameters(), lr=0.0001)  # Start with lower LR for fine-tuning
criterion = nn.CrossEntropyLoss()

print("Starting fine-tuning for 10 epochs with existing model...")
for epoch in range(10):
    model.train()
    total_loss = 0
    for images, labels in train_loader:
        images, labels = images.to("cuda"), labels.to("cuda")
        optimizer.zero_grad()
        outputs = model(images)
        loss = criterion(outputs, labels)
        loss.backward()
        optimizer.step()
        total_loss += loss.item()
    avg_loss = total_loss / len(train_loader)
    print(f"Fine-tuning Epoch {epoch+1}/10, Average Loss: {avg_loss}")

    # Validation
    model.eval()
    correct = 0
    total = 0
    with torch.no_grad():
        for images, labels in val_loader:
            images, labels = images.to("cuda"), labels.to("cuda")
            outputs = model(images)
            _, predicted = torch.max(outputs, 1)
            total += labels.size(0)
            correct += (predicted == labels).sum().item()
    accuracy = 100 * correct / total
    print(f"Validation Accuracy: {accuracy}%")

# Save the fine-tuned model
torch.save(model.state_dict(), "/content/drive/MyDrive/programs/road_damage_detection_backup/fine_tuned_best.pt")
print("Fine-tuned model saved as fine_tuned_best.pt")
```

```python
from ultralytics import YOLO
from google.colab import files
from PIL import Image

# Load the model
model_path = "/content/drive/MyDrive/programs/road_damage_detection_backup/yolov8_pothole_train/weights/best.pt"
model = YOLO(model_path)
model.to("cuda")
model.eval()
print(f"Model loaded from {model_path}")

# Upload and process the image
uploaded = files.upload()
image_path = list(uploaded.keys())[0]
image = Image.open(image_path).convert("RGB")
results = model(image)
results[0].show()

# Extract and print detection details
boxes = results[0].boxes.xyxy.cpu().numpy()
confidences = results[0].boxes.conf.cpu().numpy()
labels = results[0].boxes.cls.cpu().numpy()

detections = []
for box, conf, label in zip(boxes, confidences, labels):
    x_min, y_min, x_max, y_max = box
    severity = float((x_max - x_min) * (y_max - y_min))  # Area as severity
    detections.append({
        "x_min": float(x_min),
        "y_min": float(y_min),
        "x_max": float(x_max),
        "y_max": float(y_max),
        "confidence": float(conf),
        "label": "pothole" if label == 0 else "unknown",
        "severity": severity
    })
print("Detection Results:")
for i, det in enumerate(detections):
    print(f"Detection {i + 1}: {det}")

# Save the output
results[0].save(filename="output_image.jpg")
print("Image with detections saved as 'output_image.jpg'")

# Optional validation
try:
    val_results = model.val(data="/content/Pothole_Segmentation_YOLOv8/data.yaml")
    mAP50 = val_results.box.map50
    accuracy_percentage = mAP50 * 100
    print(f"Model Accuracy (mAP50): {accuracy_percentage:.2f}%")
except FileNotFoundError:
    print("Validation skipped: 'data.yaml' not found. Ensure the dataset is set up correctly.")
```
